# Introdução Teórica


Propriedades:

1. $E[X] =  \frac{\sum_{i=1}^{n} x_i}{n}$
2. $Var(X) = σ^2(X)$ = $\frac{\sum_{i=1}^{n} (x_i - E[x])}{n - 1}$
3. $σ(X) = \sqrt{\frac{\sum_{i=1}^{n} (x_i - E[x])}{n - 1}}$
4. $Cov(X, Y) = E[(X - E[X])\cdot(Y - E[Y])] = \frac{\sum_{i=1}^{n} (x - m_x)\cdot(y-m_y)}{n}$
5. $Corr(X,Y) = Cov(X,Y) - (E[X] \cdot E[Y])$
6. $\rho(X,Y) = \frac{Cov(X, Y)}{(σ[x] \cdot σ[Y])}$

# Importando Bibliotecas

```python
import numpy  as np
import pandas as pd

from scipy import stats
import matplotlib.pyplot as plt

import plotly.express       as px
import plotly.graph_objects as go
```

```python
SAMPLE_SIZE = 5000

np.random.seed(42)
```

# Dado 02 v.a. X e Y Normais, independentes com média zero e variância 1, gere sequências de 5.000 amostras cada para uma delas. Então:

#### a) Obtenha a $Corr(X, Y)$, a $Cov(X, Y)$ e plote o gráfico de dispersão para $XY$ (scatter de $X-Y$).

```python
# Gerando amostras de X e Y
x = stats.norm.rvs(size=SAMPLE_SIZE, loc=0, scale=1)
y = stats.norm.rvs(size=SAMPLE_SIZE, loc=0, scale=1)

# Cálculo da média
mean_x = np.mean(x)
mean_y = np.mean(y)

# Calculando a covariância entre X e Y
cov_xy = np.mean((x - mean_x) * (y - mean_y))

# Calculando a correlação entre X e Y
corr_xy = cov_xy - (mean_x * mean_y)

# Coeficiente de correlação entre X e Y
ro = cov_xy/()

# Mostrando a covariância e a correlação
print(f"Correlação (Corr): {corr_xy:.2f}")
print(f"Covariância (Cov): {cov_xy:.2f}")

# Plotando o gráfico de dispersão
plt.scatter(x, y, alpha=0.5)
plt.title(f"Gráfico de Dispersão de X-Y\nCorrelação: {corr_xy:.2f}, Covariância: {cov_xy:.2f}")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

#### b) Obtenha a $Corr(W, Z)$, a $Cov(W, Z)$, o coeficiente de correlação e plote o gráfico de dispersão para $WZ$, sendo

$W = 2\cdot X + 5$

e

$Z = Y$

```python
# Gerando amostras de X e Y
x = stats.norm.rvs(size=SAMPLE_SIZE, loc=0, scale=1)
y = stats.norm.rvs(size=SAMPLE_SIZE, loc=0, scale=1)

# Calculando W e Z com base em X e Y
W = 2 * x + 5
Z = y

# Cálculo da média
mean_w = np.mean(W)
mean_z = np.mean(Z)

# Cálculo do desvio-padrão
std_w = np.std(W)
std_z = np.std(Z)

# Calculando a covariância entre W e Z
cov_wz = np.mean((W - mean_w) * (Z - mean_z))

# Calculando a correlação entre W e Z
corr_wz = cov_wz - (mean_w * mean_z)

# Calculando o coeficiente de correlação Rô
ro = cov_wz / (std_w * std_z)

print(f'Covariância entre w e z: {cov_wz}')
print(f'Correlação entre w e z: {corr_wz}')
print(f'Coeficiente de correlação Rô: {ro}')
```

##### Plot WZ

```python
df = pd.DataFrame({'W': W, 'Z': Z})

# Plotando o gráfico de dispersão
plt.scatter(df['Z'].values, df['W'].values, alpha=0.5)
plt.title(f"Gráfico de Dispersão de W-Z\nCorrelação: {corr_wz:.2f}, \nCovariância: {cov_wz:.2f}, \nRô: {round(ro, 2)}")
plt.xlim(-15,20)
plt.ylim(-5,15)
plt.xlabel('Z')
plt.ylabel('W')
plt.grid(True)
plt.show()
```
